```python
%reset -f
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# import jtplot submodule from jupyterthemes
from jupyterthemes import jtplot
# currently installed theme will be used to
# set plot style if no arguments provided
jtplot.style()
```

```python
## import data and randomize data
diet_data = pd.read_csv('datasets/diet_data.csv')
diet_data = diet_data.sample(frac=1).reset_index(drop=True)
```

```python
## exploring data
print(diet_data.shape)
## find number null values
diet_data[diet_data.isnull().any(axis=1)].count()
```

```python
diet_data = diet_data.dropna()
```

```python
## since stone, pounds, and ounces are weight measurements, we exclude them from the training set
## since change is what we're trying to predict, we exclude them from the dataset
from sklearn.model_selection import train_test_split
X = diet_data.drop(['change','Date','Stone','Pounds','Ounces'],axis=1)
Y = diet_data['change']
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=0.2)
```

```python
x_train.sample(5)
```

```python
## standardize numeric features
## will pre-process all numeric features to have zero mean and unit variance
## recomended for sklear perceptions
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
## remember activation allows neural network to learn non linear relationships
## hidden layer size is like the neural network dimensions, this neural_network has 1 layers of neurons, 1 neuron
## this solver is an optimizer which works well on small datasets
from sklearn.neural_network import MLPRegressor
mlp_reg = MLPRegressor(activation='relu',
                      hidden_layer_sizes=(1,),
                      solver='lbfgs',
                      verbose=True,
                      max_iter=100)
mlp_reg.fit(x_train, y_train)
```

```python
y_pred = mlp_reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
mlp_reg.score(x_train, y_train)
```

```python
r2_score(y_test, y_pred)
```

```python
plt.figure(figsize=(20,10))
plt.plot(y_pred, label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('change in weight')
plt.legend()
plt.show()
```
